## Splitting data into train and test 

```python
from sklearn.model_selection import train_test_split 
x=data.drop("Price",axis=1)
y=data["Price"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=101, test_size=0.1)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## Scaling the dataset

```python
from sklearn.preprocessing import StandardScaler
scale= StandardScaler()
x_train = scale.fit_transform(x_train)
x_test = scale.transform(x_test)
```

## Creating Model - Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression(positive=False)
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print("R2 Score",r2_score(y_test,y_pred))
print("Root Mean Square Error",np.sqrt(mean_squared_error(y_test,y_pred))) 
```

```python
#Checking for model's performance on unseen data 
from sklearn.model_selection import cross_val_score
print("R2_Score_Cross_Value:",np.mean(cross_val_score(lr,x_train,y_train, scoring = "r2", cv = 5)))
```

## Checking if residuals are normal distributed

```python
residuals=y_test-y_pred
```

```python
sns.kdeplot(residuals)
plt.show()
```
